```python
# Task:1 - Prediction using Supervised ML | The Sparks Foundation
# Developer: Yatharth Chauhan

# Predict the percentage of an student based on the no. of study hours.
# This is a simple linear regression task as it involves just 2 variables.
# You can use R, Python, SAS Enterprise Miner or any other tool
# Data can be found at http://bit.ly/w-data
# What will be predicted score if a student studies for 9.25 hrs/ day?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

url = "http://bit.ly/w-data"
data = pd.read_csv(url)
data.head()

X = data['Hours'].values.reshape(-1, 1)
y = data['Scores'].values

model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hours', y='Scores', data=data)
plt.plot(data['Hours'], predictions, color='red', linewidth=2)
plt.title('Study Hours vs. Percentage Scores')
plt.xlabel('Study Hours')
plt.ylabel('Percentage Scores')
plt.show()

hours= float(input("Enter no. of study hours: "))
prediction = model.predict([[hours]])

print("Predicted score if a student studies for {} hrs/day: {}".format(hours, prediction[0]))

print("\nCode with Yatharth Chauhan❤️")
```
